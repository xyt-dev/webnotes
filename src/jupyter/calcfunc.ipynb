```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def xln(x):
  return x * np.log(x)
x = np.linspace(-1, 2, 1000)
y = xln(x)
plt.figure()
plt.plot(x, y)
plt.grid(True)
plt.title("$f(x)=xlnx$")
plt.show()
```

```python
x = np.linspace(-0.1, 0.1, 1000)
y = x**2 * np.sin(1/x)
plt.figure()
plt.plot(x, y)
plt.grid(True)
plt.title("$f(x)=x^2sin(1/x)$")
plt.show()
```
